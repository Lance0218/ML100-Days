# 課堂重點筆記  

### bagging:
* 用途: 用於很強的model，避免overfitting  
* 說明: 將資料切分，每一次的training data都只用到部分，剩下的做驗證，可以達到類似validation的效果
* e.g.: out-of-bag

### boosting:
* 用途: 用於弱的model，幫助fit model
* 說明: 找出一個弱分類器，在找下一個與其互補的弱分類器，一直找下去再集合起來
* e.g.: AdaBoost、Gradient Boosting
